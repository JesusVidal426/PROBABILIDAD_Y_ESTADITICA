```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el conjunto de datos

datos = pd.read_csv('/content/grupo1.csv')
print(datos.head())
print(datos.describe())
plt.hist(datos['Calificacion'], bins=10, color='blue', edgecolor='black')
plt.xlabel('Calificaciones')
plt.ylabel('Frecuencia')
plt.title('Histograma de las Calificaciones')
plt.show()
```

```python
# MODELO DE REGRESION LINEAL
from sklearn.linear_model import LinearRegression
from scipy.stats import pearsonr
horas_estudio = datos[['HorasEstudio']]
calificaciones = datos['Calificacion']
modelo = LinearRegression()
modelo.fit(horas_estudio, calificaciones)
pendiente = modelo.coef_[0]
intercepto = modelo.intercept_
print("Ecuación de la línea de regresión:")
print(f"Calificaciones = {pendiente:.2f} * Horas de Estudio + {intercepto:.2f}")
```

```python
# VISUALIZACION DE RESULTADOS Y CALCULO DE LA CORRELACION
import numpy as np
plt.scatter(datos['HorasEstudio'], datos['Calificacion'], color='purple', label='Datos')
plt.plot(datos[['HorasEstudio']], modelo.predict(horas_estudio), color='black', label='Línea de regresión')
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificaciones')
plt.title('Regresión Lineal: Horas de Estudio vs Calificaciones')
plt.legend()
plt.show()
coef_corr, _ = pearsonr(datos['HorasEstudio'], datos['Calificacion'])
print("El coeficiente de correlación de Pearson es:", coef_corr)
```

```python
# Importar bibliotecas necesarias
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Suponiendo que 'datos' es un DataFrame de pandas, 'modelo' es un modelo de scikit-learn
# y 'horas_estudio' es una matriz de características para la predicción
import pandas as pd
from sklearn.linear_model import LinearRegression

# Datos de ejemplo
datos = pd.DataFrame({
    'HorasEstudio': [1, 2, 3, 4, 5],
    'Calificacion': [2, 4, 5, 4, 5]
})

# Suponiendo que 'modelo' ya está entrenado
modelo = LinearRegression()
horas_estudio = datos[['HorasEstudio']]  # Característica para predicción
modelo.fit(horas_estudio, datos['Calificacion'])

# Visualización de resultados y cálculo de la correlación
plt.scatter(datos['HorasEstudio'], datos['Calificacion'], color='lightgreen', label='Datos')
plt.plot(horas_estudio, modelo.predict(horas_estudio), color='brown', label='Línea de regresión')
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificaciones')
plt.title('Regresión Lineal: Horas de Estudio contra Calificaciones')
plt.legend()
plt.show()

# Calcular el coeficiente de correlación de Pearson
coef_corr, _ = pearsonr(datos['HorasEstudio'], datos['Calificacion'])
print("El coeficiente de correlación de Pearson es:", coef_corr)
```

```python
# MODELO DE REGRESION LINEAL
from sklearn.linear_model import LinearRegression
from scipy.stats import pearsonr

# Preparar los datos
horas_estudio = datos[['HorasEstudio']]
calificaciones = datos['Calificacion']

# Crear el modelo de regresión lineal
modelo = LinearRegression()
modelo.fit(horas_estudio, calificaciones)

# Obtener la pendiente e intercepto
pendiente = modelo.coef_[0]
intercepto = modelo.intercept_
print("Ecuación de la línea de regresión:")
print(f"Calificaciones = {pendiente:.2f} * Horas de Estudio + {intercepto:.2f}")

```

```python
# VISUALIZACION DE RESULTADOS Y CALCULO DE LA CORRELACION
import numpy as np
plt.scatter(datos['HorasEstudio'], datos['Calificacion'], color='green', label='Datos')
plt.plot(datos[['HorasEstudio']], modelo.predict(horas_estudio), color='orange', label='Línea de regresión')
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificaciones')
plt.title('Regresión Lineal: Horas de Estudio vs Calificaciones')
plt.legend()
plt.show()
coef_corr, _ = pearsonr(datos['HorasEstudio'], datos['Calificacion'])
print("El coeficiente de correlación de Pearson es:", coef_corr)
```

1. Interpretación del Coeficiente de Regresión
El coeficiente de regresión indica la pendiente de la línea de regresión y nos muestra cuánto cambia la calificación promedio por cada hora adicional de estudio. Por ejemplo, si el coeficiente de regresión es 1.5, esperamos que por cada hora extra de estudio, la calificación aumente en 1.5 puntos.

2. Interpretación del Coeficiente de Correlación
El coeficiente de correlación de Pearson (
𝑟
r) varía entre -1 y 1. Un valor cercano a 1 indica una fuerte correlación positiva, sugiriendo que a mayor tiempo de estudio, mayores son las calificaciones. Un valor cercano a -1 indica una fuerte correlación negativa, sugiriendo que a mayor tiempo de estudio, menores son las calificaciones. Un valor cercano a 0 indica que no hay una relación lineal significativa. Para este análisis, un valor más cercano a 1 sería ideal, ya que sugiere que el aumento en el tiempo de estudio está asociado con mejores calificaciones.

3. Significancia Estadística del Modelo de Regresión Lineal
El modelo de regresión lineal es estadísticamente significativo cuando permite predecir con precisión los valores. Esto significa que podemos usar el modelo para estimar, por ejemplo, si un estudiante que estudia 20 horas puede obtener una calificación aprobatoria, o si un estudiante que estudia pocas horas probablemente obtendrá una calificación baja.

4. Confiabilidad del Modelo de Regresión
La confiabilidad del modelo de regresión puede evaluarse usando el coeficiente de determinación (R²). Aunque el modelo puede no encajar perfectamente en todos los casos, un R² alto indica que el modelo explica bien la variabilidad de las calificaciones basada en el tiempo de estudio. En general, los análisis muestran que hay una relación positiva y significativa entre el tiempo de estudio y las calificaciones de los estudiantes, lo que sugiere que el modelo es confiable para hacer predicciones basadas en el tiempo de estudio.

Conclusiones Generales
En términos generales, se observa que a medida que los estudiantes aumentan el tiempo dedicado al estudio, sus calificaciones tienden a mejorar. Los modelos de regresión lineal desarrollados para los tres grupos de datos indican que el tiempo de estudio es un predictor sólido de las calificaciones. Esto se refleja en un alto valor de R² y un bajo p-valor, lo que sugiere que los resultados son estadísticamente significativos y confiables para predecir el rendimiento académico basado en el tiempo de estudio.

Sin embargo, es importante destacar que el grupo 1 muestra una dispersión mayor en los datos, lo que complica la precisión de los cálculos y la interpretación de la regresión lineal en este caso específico. A pesar de esta mayor dispersión, el modelo aún proporciona información útil sobre la relación entre el tiempo de estudio y las calificaciones, aunque pueda requerir un análisis más detallado y cauteloso.
